```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")

pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv("../data/raw/loan_data.csv")

df.head()
```

```python
df = pd.read_csv("../data/raw/train.csv")

df.head()
```

```python
print("Shape :", df.shape)

df.info()
```

```python
# Missing Values
df.isnull().sum()
```

```python
# Missing Values Visualization
plt.figure(figsize=(10,5))
sns.heatmap(df.isnull(), cbar=False)
plt.show()
```

```python
# Target Variable Distribution
sns.countplot(
    x='Loan_Status',
    data=df
)

plt.title("Loan Approval Distribution")
plt.show()
```

```python
# Numerical Summary
df.describe()
```

```python
#Correlation Analysis

numeric_df = df.select_dtypes(include=np.number)

plt.figure(figsize=(10,6))
sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap="coolwarm"
)
plt.show()
```

```python
# Income Distribution
sns.histplot(
    df["ApplicantIncome"],
    kde=True
)

plt.show()
```

```python
# Loan Amount Distribution
sns.histplot(
    df["LoanAmount"],
    kde=True
)

plt.show()
```

```python
# redit History vs Loan Status
sns.countplot(
    x='Credit_History',
    hue='Loan_Status',
    data=df
)

plt.show()
```

```python
# redit History vs Loan Status
sns.countplot(
    x='Credit_History',
    hue='Loan_Status',
    data=df
)

plt.show()
```
