```python
import pandas as pd

df = pd.read_csv("../data/raw/train.csv")

df.head()
```

```python
cat_cols = [
    'Gender',
    'Married',
    'Dependents',
    'Self_Employed',
    'Education',
    'Property_Area'
]

num_cols = [
    'ApplicantIncome',
    'CoapplicantIncome',
    'LoanAmount',
    'Loan_Amount_Term',
    'Credit_History'
]
```

```python
for col in cat_cols:
    df[col]=df[col].fillna(
        df[col].mode()[0]
    )

for col in num_cols:
    df[col]=df[col].fillna(
        df[col].median()
    )
```

```python
#Feature Creation
df["TotalIncome"] = (
    df["ApplicantIncome"] +
    df["CoapplicantIncome"]
)
```

```python
df["IncomePerPerson"] = (
    df["TotalIncome"] /
    (df["Dependents"]
      .replace("3+",3)
      .replace("0",1)
      .astype(int))
)
```

```python
df["LoanIncomeRatio"] = (
    df["LoanAmount"] /
    df["TotalIncome"]
)
```

```python
# Label Encoding
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in df.columns:
    if df[col].dtype == "str":
        df[col] = le.fit_transform(df[col])
```
